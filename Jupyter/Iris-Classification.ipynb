```python
data_bias = np.ones((150, 1))
```

```python
data_y_raw = data_3[:, -1]
```

```python
data_y_raw = data_y_raw.reshape((len(data_y_raw), 1))
```

```python
data_4 = data_3[:,:-1]
```

```python
data = np.column_stack((data_bias, data_4))
```

```python
def sigmoid(inX):
    return 1 / (1 + np.exp(-inX))
```

```python
def cost_func(inData, inTheta, inY):
    #m = 1 / len(inY)
    m = 1 / np.sqrt(len(inY))
    pred = sigmoid(inData.dot(inTheta))
    cost = m * np.sum((-inY * np.log(pred) - (1 - inY) * np.log(1 - pred)))
    return cost
```

```python
def gradient_func(inData, inTheta, inTarget):
    #m = 1 / len(inTarget)
    m = 1 / np.sqrt(len(inTarget))
    pred = sigmoid(inData.dot(inTheta))
    gradient = m * inData.T.dot(pred - inTarget)
    #gradient = sigmoid(inData.T.dot(pred - inTarget))
    return gradient
```

```python
data_y_1 =data_y_raw
for i in range(len(data_y_1)):
    if data_y_1[i] == 1:
        data_y_1[i] = 0
    else:
        data_y_1[i] = 1
```

```python
cost = []
theta = np.zeros((5, 1))
maxitr = 5000
alpha = 1 / maxitr
for i in range(maxitr):
    theta -= alpha * gradient_func(data, theta, data_y_1)
    cost.append(cost_func(data, theta, data_y_1))
```

```python
pp.plot(cost)
```
